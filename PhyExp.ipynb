```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
FFTCu = pd.read_csv("80-20Cu.csv")
FFTCN = pd.read_csv("C-Nigel.csv")
FFTPB = pd.read_csv("phrospher bronze.csv")
```

```python
FFTCu["U/V"]
```

```python
peakCu,_ = find_peaks(FFTCu["U/V"], height=32, prominence=30)
```

```python
peakCN,_ = find_peaks(FFTCN["U/V"], height=32, prominence=40)
```

```python
peakPB,_ = find_peaks(FFTPB["U/V"], height=40, prominence=10)
```

```python
df = FFTCu["U/V"]
pk = peakCu
plt.subplot(2, 2, 1)
plt.plot(pk, df[pk], "or"); plt.plot(df); plt.legend(['Copper'])
```

```python
df = FFTCN["U/V"]
pk = peakCN
plt.subplot(2, 2, 1)
plt.plot(pk, df[pk], "or"); plt.plot(df); plt.legend(['Carbon-Nickel'])
```

```python
df = FFTPB["U/V"]
pk = peakPB
plt.subplot(2, 2, 1)
plt.plot(pk, df[pk], "or"); plt.plot(df); plt.legend(['Phosphor-Bronze'])
```
